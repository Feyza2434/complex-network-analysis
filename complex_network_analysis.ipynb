```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import pandas as pd
```

```python
games=pd.read_csv("mlb_games.csv", encoding='utf-16')
teams=pd.read_csv("mlb_teams.csv", encoding='utf-16')
```

```python
edges_undirected = games[["home_team", "away_team"]].copy()

# Tekrar eden kenarları sildik.
edges_undirected = edges_undirected.drop_duplicates()
# Yönsüz graf
G = nx.Graph()
G.add_edges_from(edges_undirected.itertuples(index=False, name=None))

plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=True, node_size=300, font_size=8,node_color='pink')
plt.title("Yönsüz Ağ")
plt.show()
```

```python
# Yönlü graf
edges_directed = games[["home_team", "away_team"]].copy()

G_directed=nx.DiGraph()
G_directed.add_edges_from(edges_directed.itertuples(index=False, name=None))
plt.figure(figsize=(10,8))
nx.draw(G_directed,with_labels=True,node_size=200,arrows=True,node_color='lightcoral',font_size=8)
```

```python
# Derece hesaplama
degrees = dict(G.degree())

print("Düğümlerin dereceleri:")
for node, deg in degrees.items():
    print(f"{node}: {deg}")
```

```python
# Ortalama derece hesaplama
avg_degree=sum(degrees.values())/G.number_of_nodes()
print(avg_degree)
```

```python
# Histogram grafiği
degree_list = [G.degree(n) for n in G]

plt.figure(figsize=(8, 6))
plt.hist(degree_list)
plt.xlabel("Derece")
plt.ylabel("Frekans")
plt.title("Histogram")
plt.show()

```

```python
deg_centrality = nx.degree_centrality(G)

for node, val in deg_centrality.items():
    print(f"{node}: {val}")
```

```python
closeness_centrality = nx.closeness_centrality(G)

for node, val in closeness_centrality.items():
    print(f"{node}: {val}")
```

```python
betweenness_centrality = nx.betweenness_centrality(G)

for node, val in betweenness_centrality.items():
    print(f"{node}: {val}")
```

```python
from networkx.algorithms.centrality import eigenvector
eigenvector_centrality = nx.eigenvector_centrality(G, max_iter=1000)

for node, val in eigenvector_centrality.items():
    print(f"{node}: {val}")
```

```python
katz_centrality = nx.katz_centrality(G, alpha=0.01, beta=1.0, max_iter=5000)

for node, val in katz_centrality.items():
    print(f"{node}: {val}")
```

```python
pagerank_centrality = nx.pagerank(G, alpha=0.85)

for node, val in pagerank_centrality.items():
    print(f"{node}: {val}")
```

```python
from networkx.algorithms.community import greedy_modularity_communities
from networkx.algorithms.community.quality import modularity

communities = greedy_modularity_communities(G)
modularity = modularity(G, communities)
print("Modularity:", modularity)
```

```python
# Density hesaplama
density=nx.density(G)
print("Density:",density)
```

```python
avg_clustering=nx.average_clustering(G)
print("Average Clustering:",avg_clustering)
```

```python
# Çap
diameter=nx.diameter(G)
print("Diameter:",diameter)
```

```python
# Derece dağılımı
degree_list = [G.degree(n) for n in G]

for d in sorted(set(degree_list)):
    print(d, ":", degree_list.count(d))
```
